# Categorical plots #

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/sanakulov/Private/ML/ml/05-Seaborn/StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.set_theme()
sns.set_style('darkgrid')
sns.set_context("paper")
plt.figure(figsize=(6,3), dpi=100)
sns.boxplot(data=df,x='writing score', y='parental level of education', hue='test preparation course', palette="Set2").set_title("Fig 1")
plt.legend(loc=(1.05,0.85), title='test preparation course')
```

```python
sns.set_theme()
sns.set_style('darkgrid')
sns.set_context("paper")
plt.figure(figsize=(6,5), dpi=100)
sns.violinplot(data=df,x='writing score', y='parental level of education', palette="Set2", inner='stick').set_title("Fig 1")
plt.legend(loc=(1.05,0.85), title='test preparation course')
```

```python
sns.set_theme()
sns.set_style('darkgrid')
sns.set_context("paper")
plt.figure(figsize=(6,5), dpi=100)
sns.violinplot(data=df,x='writing score', y='parental level of education', bw_method=0.1, palette="Set2").set_title("Fig 1") #bw = bandwigth   ()
plt.legend(loc=(1.05,0.85), title='test preparation course')
```

```python
plt.figure(figsize=(6,2), dpi=100)
sns.swarmplot(data=df,x="math score", y="gender", hue="test preparation course", size=2, dodge=True)
plt.legend(loc=(1.1,0.8), title="test preparation course")
```

```python
plt.figure(figsize=(6,3), dpi=100)
sns.boxenplot(data=df,x="math score", y="test preparation course", hue="gender")
plt.legend(loc=(1.1,0.8), title="gender")
```

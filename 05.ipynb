```python
plt.figure(figsize=(5,2), dpi=100)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.set(style='darkgrid') #  + whitegrid + white + dark + ticks 
#set style - ,       -  ,   kernel
sns.displot(x='salary', data=df, bins=11, color="red", edgecolor="black", lw=4, kde=True, rug=True, height=3, aspect=1.5) #displot -    - -  height  aspect
```

```python
plt.figure(figsize=(5,2), dpi=100)
sns.kdeplot(data=df,x="salary", fill="blue")
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200) #     
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x="age")
```

```python
sns.displot(data=sample_ages, x="age", rug=True, bins=30, kde=True) # kde    ,      
```

```python
sns.set_context("paper", rc={"font.size":8,"axes.titlesize":8,"axes.labelsize":8})
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,3))
g1 = sns.kdeplot(data=sample_ages,x="age", ax=ax[0]) #
g2 = sns.kdeplot(data=sample_ages,x="age", clip=[0,100], ax=ax[1], color="red", bw_adjust=0.6, fill=True) #, bw = bandwith -    -      ,       
fig.figure.suptitle("comparison", size=16)
g1.set_title("classic", size=10)
g2.set_title("clipped", size=10)
fig.tight_layout()
#    ?    age=20,      ( 0.009)
```

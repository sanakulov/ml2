```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2, figsize=(4,4), dpi=100)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][1].plot(a,b)
axes[1][0].plot(x,y)

axes[0][0].set_ylabel('Y label')
axes[0][1].set_ylabel('Y label')
axes[1][1].set_ylabel('Y label')
axes[1][0].set_ylabel('Y label')

axes[0][0].set_xlabel('X label')
axes[0][1].set_xlabel('X label')
axes[1][1].set_xlabel('X label')
axes[1][0].set_xlabel('X label')

axes[0][0].set_title('Title')
axes[0][1].set_title('Title')
axes[1][1].set_title('Title')
axes[1][0].set_title('Title')

plt.tight_layout()

fig.suptitle('Super Title')

# fig.set_figwidth(4)
# fig.set_figheight(4)

fig.savefig('newsubplots.png', bbox_inches='tight')

# fig.subplots_adjust(wspace=0.3, hspace=0.4)
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='x vs x')

ax.plot(x,x**2,label='x vs x^2')

# ax.legend(loc='upper left') 
ax.legend(loc=(1.1,0.5)) 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

plt.xlabel(" x")
plt.ylabel(" y")
plt.title("test_unit_1")
plt.xlim(0,10)
plt.ylim(-5,20)

# ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
# ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
# ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
# ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes

ax.grid(True)
ax.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)

ax.text(0.3,0.3, r"$y=x^2$", fontsize = 20, color="blue") #$   $
ax.text(8,6, r" ", fontsize = 20, color="red")

ax.plot(x,x, color='#90a1d0', label="blue line", lw=3) #lw == linewidth
ax.plot(x*2,x*3, color='#ec6064', label="red line", ls='--')
ax.plot(x**1.3, color='#b7ce77', label="green line", marker="o", ms=10, markerfacecolor="red", markeredgewidth="4", markeredgecolor="orange") #ms == markersize
ax.legend(loc=(1.05,0.9))
```

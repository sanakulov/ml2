```python
# Считаем, сколько нулловых значений по колонкам
df.isnull().sum()
```

```python
# Но полезнее считать % данных
df.isnull().sum()*100 / len(df)
```

```python
def percent_missing(my_df):
    result = my_df.isnull().sum()*100 / len(my_df)
    return result[result>0].sort_values()
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
plt.figure(figsize=(7,3),dpi=200)
ax = sns.barplot(x=percent_nan.index,y=percent_nan, hue=percent_nan).set_title('Missing Data')
plt.xticks(rotation=90)
plt.ylim(0,1) # посмотрим, что там по 1% строк
```

```python
# Считаем, сколько нулловых значений по колонкам
df.isnull().sum()
```

```python
# Но полезнее считать % данных
df.isnull().sum()*100 / len(df)
```

```python
def percent_missing(my_df):
    result = my_df.isnull().sum()*100 / len(my_df)
    return result[result>0].sort_values()
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
plt.figure(figsize=(7,3),dpi=200)
ax = sns.barplot(x=percent_nan.index,y=percent_nan, hue=percent_nan).set_title('Missing Data')
plt.xticks(rotation=90)
plt.ylim(0,1) # посмотрим, что там по 1% строк
```

```python
percent_nan[percent_nan<1] # отбираем признаки, у кот % заполнения меньше 1%
```

```python
# Стратегия такая: желательно оставлять как можно больше данных - полезно для обучения модели. Но тут всего несколько строк, дешевле их дропать
# Имеет смысл начинать с тех строк, которые имеют наименьший % отсутствующих строк


df = df.dropna(axis=0, subset=['Electrical', 'Garage Area']) # ЛФ чистки датафрейма налловых значений только по ряду параметров
```

```python
percent_nan = percent_missing(df)
```

```python
# Получается, что null в колонках связанных с подвалом обозначает отсутствие подвала как такогового. Можем записать в числовые колонки нули, а в текстовые колонки none (в площади подвалов нули)
# ЛФ замены null налл на значения

## для числовых колонок
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF','Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)
## для текстовых колонок
bsmt_str_cols =  ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_str_cols] = df[bsmt_str_cols].fillna('None')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/Users/sanakulov/Private/ML/ml/05-Seaborn/StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(4,4), dpi=50)
sns.jointplot(data=df,x="math score",y="reading score", kind="scatter",hue="gender") #через kind настраивается вид jointplot
```

```python
plt.figure(figsize=(2,2), dpi=200)
sns.pairplot(data=df,hue="gender",corner=True)
```

```python
sns.catplot(data=df,x="gender",y="math score",kind="box",col="lunch",row="test preparation course")
```

```python
g=sns.PairGrid(data=df, hue="gender")
g=g.map_upper(sns.scatterplot)
g=g.map_lower(sns.kdeplot)
g=g.map_diag(sns.kdeplot)
g=g.add_legend()
```
